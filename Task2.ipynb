```python
import pandas as pd

# add data set
df = pd.read_csv('cleaned_dataset.csv')

df.head()
```

```python
#Dataset shap and info
print(df.shape)
df.info

#statistics
df.describe()

# check mising value
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set (style="whitegrid")
```

```python
#Distribution of Likes
plt.figure(figsize=(8,5))
sns.histplot(df['Likes'], bins=15, kde=True, color='skyblue')
plt.title("Distribution of Likes")
plt.xlabel("Likes")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
#Likes by Country
plt.figure(figsize=(10, 10))
#sns.boxplot(x='Country', y='Likes', data=df, palette='Set2')
sns.boxplot(data=df, x='Platform', y='Likes')
plt.title("Likes Distribution Country wise")
plt.show()
```

```python
#Likes vs Hour colored by Platform
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Hour', y='Likes', hue='Platform', data=df, palette='Set1', s=100)
plt.title("Scatter Plot of Likes by Hour across Platforms")
plt.xlabel("Hour of Day")
plt.ylabel("Likes")
plt.tight_layout()
plt.show()
```

```python
#descriptive statistics
print("Statistics for Numerical Features:\n")
print(df.describe())

#Extract only numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64'])

#Compute the correlation matrix
correlation_matrix = numerical_features.corr()

print("\nCorrelation Matrix:\n")
print(correlation_matrix)

# Visualize correlation matrix with heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, linecolor='white')
plt.title("Correlation Matrix Heatmap of Numerical Features")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
