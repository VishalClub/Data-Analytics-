```python
#describe bar plote
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('cleaned_dataset.csv')

# Set Seaborn style
sns.set(style="whitegrid")
```

```python
# Ratweets by Platform
plt.figure(figsize=(8, 5))
platform_retweets = df.groupby('Platform')['Retweets'].sum().reset_index()
sns.barplot(x='Platform', y='Retweets', data=platform_retweets, palette='Set3')
plt.title("Total Retweets by Platform")
plt.xlabel("Platform")
plt.ylabel("Total Retweets")
plt.tight_layout()
plt.show()
```

```python
# post count use by Country
plt.figure(figsize=(50, 20))
country_counts = df['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Post Count']
sns.barplot(x='Country', y='Post Count', data=country_counts, palette='coolwarm')
plt.title("Number of Posts by Country")
plt.xlabel("Country")
plt.ylabel("Post Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# use a line chart to count parday post
plt.figure(figsize=(8, 5))
posts_by_day = df['Day'].value_counts().sort_index().reset_index()
posts_by_day.columns = ['Day', 'Post Count']
sns.lineplot(x='Day', y='Post Count', data=posts_by_day, marker='o', color='purple')
plt.title("Post Count by Day of Month")
plt.xlabel("Day")
plt.ylabel("Number of Posts")
plt.xticks(range(1, 32))
plt.tight_layout()
plt.show()
```

```python
#Use of scatter plots to Likes vs Hour colored by Platform
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df,x='Hour',y='Likes',hue='Platform',palette='Dark2',s=100,edgecolor='w')
plt.title("Scatter Plot of Likes vs Hour by Platform")
plt.xlabel("Hour of Day")
plt.ylabel("Likes")
plt.tight_layout()
plt.show()
```

```python
# Customize plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=df,x='Hour',y='Likes',hue='Platform',palette='Set1',s=100,edgecolor='white')

# Customize title and labels
plt.title("User Engagement:", fontsize=14, fontweight='bold')
plt.xlabel("Hour of Day (0 to 23)", fontsize=12)
plt.ylabel("Number of Likes", fontsize=12)

# Customize legend
plt.legend(title='Social Media Platform', title_fontsize=11, fontsize=10, loc='upper right', frameon=True)

# Customize tick labels
plt.xticks(range(0, 24), rotation=0)
plt.yticks(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)

# Apply tight layout
plt.tight_layout()

# Show the plot
plt.show()
```
